# Prepocessing

```python
d = {
    'zero' : 0,
    'one': 1,
    'two': 2,
    'three' : 3,
    'four' : 4,
    'five' : 5,
    'six' : 6,
    'seven' : 7,
    'eight': 8,
    'nine' : 9,
    'ten': 10,
    'eleven': 11
}

data['experience'] = data['experience'].map(d)

data['experience']
```

```python
data['experience'].fillna(0, inplace = True)
data['experience']
```

```python
#test score null value handle
test_score_median = data['test_score(out of 10)'].median()

data['test_score(out of 10)'].fillna(test_score_median, inplace = True)
data['test_score(out of 10)']
```

```python
data.isnull().any()
```

#Train the model

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['experience', 'test_score(out of 10)','interview_score(out of 10)']]
y = data['salary($)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print('Intercept: ', model.intercept_)
print('Coef1: ', model.coef_[0])
print('Coef2: ', model.coef_[1])
print('Coef3: ', model.coef_[2])
```

#evaluation

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print("r^2 : ", r2)
```
